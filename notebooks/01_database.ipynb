### 🏗️ Passo 2: Definindo nossos Modelos

Aqui definimos as tabelas do banco de dados:
1. **Broker**: O usuário do SaaS (você).
2. **Client**: Seus segurados.
3. **Policy**: As apólices de carros e motos.

```python
from sqlalchemy import Column, Integer, String, ForeignKey, Date, Float
from sqlalchemy.orm import relationship, declarative_base

Base = declarative_base()

class Broker(Base):
    __tablename__ = 'brokers'
    id = Column(Integer, primary_key=True)
    name = Column(String, nullable=False)
    email = Column(String, unique=True, nullable=False)
    clients = relationship("Client", back_populates="broker")

class Client(Base):
    __tablename__ = 'clients'
    id = Column(Integer, primary_key=True)
    name = Column(String, nullable=False)
    broker_id = Column(Integer, ForeignKey('brokers.id'))
    broker = relationship("Broker", back_populates="clients")
    policies = relationship("Policy", back_populates="client")

class Policy(Base):
    __tablename__ = 'policies'
    id = Column(Integer, primary_key=True)
    policy_number = Column(String)
    expiration_date = Column(Date)
    client_id = Column(Integer, ForeignKey('clients.id'))
    client = relationship("Client", back_populates="policies")

print("Modelos definidos com sucesso!")
```

### 🚀 Passo 3: Criando o Banco de Dados

Agora vamos inicializar o arquivo `insurance.db` e criar as tabelas.

```python
from sqlalchemy import create_engine

engine = create_engine("sqlite:///./insurance.db")
Base.metadata.create_all(bind=engine)

print("Banco de dados 'insurance.db' criado com sucesso!")
```
