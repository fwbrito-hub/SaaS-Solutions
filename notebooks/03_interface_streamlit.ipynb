```python
# Exemplo rápido de como o Streamlit funciona:
import streamlit as st

st.title("Olá, corretor!")
nome = st.text_input("Qual o seu nome?")
if nome:
    st.write(f"Bem-vindo ao futuro dos seguros, {nome}!")
```
