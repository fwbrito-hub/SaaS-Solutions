### 3. Criando o Agente Inteligente
Agora vamos instanciar o nosso agente que sabe ler SQL.

```python
from agno.agent import Agent
from agno.models.groq import Groq
from agno.tools.sql import SQLTools

agent = Agent(
    model=Groq(id="llama-3.3-70b-versatile"),
    tools=[SQLTools(db_url="sqlite:///insurance.db")],
    markdown=True,
    instructions=[
        "Você é um assistente de corretora de seguros.",
        "Sua missão é ajudar o corretor a analisar dados de clientes e apólices.",
    ]
)

agent.print_response("Quais tabelas existem no meu banco de dados?", stream=True)
```
